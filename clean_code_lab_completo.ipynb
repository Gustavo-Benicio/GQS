## Exercício 7 — Simplicidade

## Exercício 9 — Testes

```
# Código Sujo
def eh_par(n):
    if n % 2 == 0:
        return True
    else:
        return False

# Código Limpo
def eh_par(n):
    return n % 2 == 0

def testar():
    assert eh_par(2)
    assert not eh_par(3)
    print("OK")

testar()
```

## Exercício 10 — Média, Máx, Mín

```
# Código Sujo
lista=[1,2,3,4,5]
media=sum(lista)/len(lista)
print(media,max(lista),min(lista))

# Código Limpo
def analisar(v):
    if not v: return {"erro": "lista vazia"}
    return {
        "média": sum(v)/len(v),
        "máximo": max(v),
        "mínimo": min(v)
    }

analisar([10,20,30,40])
```

## Exercício 11 — Sistema Legado → Refatoração Completa

```
# Código Sujo (Sistema Legado)
def executar_sistema():
    print("Iniciando sistema...")
    opcao = input("1- Criar usuário | 2- Deletar usuário:")

    if opcao == "1":
        nome = input("Nome: ")
        email = input("Email: ")
        print("Salvando no banco...")
        print("Enviando email de boas-vindas para", email)
        print("Enviando push notification...")
        print("Log: Usuário criado")

    elif opcao == "2":
        email = input("Email do usuário a deletar: ")
        print("Deletando do banco...")
        print("Enviando email de remoção para", email)
        print("Enviando push notification...")
        print("Log: Usuário deletado")

    else:
        print("Opção inválida")

# Código Limpo (Refatorado)
class NotificationService:
    def enviar_email(self, destino, assunto):
        print(f"[EMAIL] {destino} | {assunto}")

    def push(self, msg):
        print(f"[PUSH] {msg}")

    def log(self, msg):
        print(f"[LOG] {msg}")

class UserService:
    def __init__(self, notifier):
        self.notifier = notifier

    def criar_usuario(self, nome, email):
        print("Salvando usuário...")
        self.notifier.enviar_email(email, "Bem-vindo!")
        self.notifier.push("Usuário criado")
        self.notifier.log(f"Usuário criado: {email}")

    def deletar_usuario(self, email):
        print("Deletando usuário...")
        self.notifier.enviar_email(email, "Conta removida")
        self.notifier.push("Usuário removido")
        self.notifier.log(f"Usuário deletado: {email}")

class App:
    def __init__(self):
        self.notifier = NotificationService()
        self.user_service = UserService(self.notifier)

    def executar(self):
        print("1 - Criar usuário")
        print("2 - Deletar usuário")
        opcao = input("Escolha: ")

        if opcao == "1":
            nome = input("Nome: ")
            email = input("Email: ")
            self.user_service.criar_usuario(nome, email)

        elif opcao == "2":
            email = input("Email: ")
            self.user_service.deletar_usuario(email)

        else:
            print("Opção inválida")

# App().executar()  # Comentado para não travar o notebook
```
